# **EDA**

 **Univarient Analysis**

```python
sns.set(style="darkgrid")
plt.figure(figsize=(8,6))
plt.plot(old_car_data['manufacture'], color='r',alpha=0.5)
plt.title(f'{num_columns} of the old car', size=20)
plt.xlabel('x-axis')
plt.ylabel('yaxis')
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(8,6))
plt.plot(old_car_data['manufacture'][:20], color='r', alpha=0.5)
plt.title(f'{num_columns} of the car', size=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot()
```

```python
cat_columns
```

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  plt.plot(old_car_data[i],color='r', alpha=0.5)
  plt.title(f'{i} of the car')
  plt.xlabel('x-axis')
  plt.ylabel('y-axis')
  plt.plot()
```

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  plt.plot(old_car_data[i][:20],color='r', alpha=0.5)
  plt.title(f'{i} of the car')
  plt.xlabel('x-axis')
  plt.ylabel('y-axis')
  plt.plot()
```

# **BarPlot**

```python
for i in num_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  old_car_data[i].value_counts().plot(kind='bar')
  plt.title(f'{i} of the old car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

# **Categorical Columns**

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  old_car_data[i].value_counts().plot(kind='bar', color='r')
  plt.title(f'{i} of the car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  old_car_data[i].value_counts()[:15].plot(kind='bar', color='r')
  plt.title(f'{i} of the car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

# **Histogram**

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  old_car_data[i].value_counts().plot(kind='hist', color='r')
  plt.title(f'{i} of the car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  old_car_data[i].value_counts()[:20].plot(kind='bar', color='r')
  plt.title(f'{i} of the car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

```python
#histogram works better with numerical data
for i in num_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  old_car_data[i].value_counts()[:15].plot(kind='hist', color='r')#used for continuous data
  plt.title(f'{i} of the car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

# **Count Plot**

```python
for i in num_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  sns.countplot(data=old_car_data, x=i)
  plt.title(f'{i} of the old car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  sns.countplot(data=old_car_data, x=i)
  plt.title(f'{i} of the old car', size=20)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```

```python
for i in cat_columns:
  sns.set(style="darkgrid")
  plt.figure(figsize=(8,6))
  sns.countplot(data= old_car_data[:20], x=i)
  plt.title(f'{i} of the old car')
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.show()
```
